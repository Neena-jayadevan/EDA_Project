## 3.Which payment method is the most popular, and does it vary by branch or customer type?

```python
# Most popular payment method overall
print(data['Payment'].value_counts())

# Payment method popularity by branch and customer type
payment_by_branch = data.groupby(['Branch', 'Payment']).size().unstack()
payment_by_customer = data.groupby(['Customer type', 'Payment']).size().unstack()

# Plot overall payment method popularity
data['Payment'].value_counts().plot(kind='barh', color='green', title='Payment Method Popularity')
plt.ylabel("Count")
plt.show()

# Plot payment method popularity by branch
payment_by_branch.plot(kind='bar',stacked=True, figsize=(7, 7), title='Payment Method by Branch')
plt.ylabel("Count")
plt.show()

# Plot payment method popularity by customer type
payment_by_customer.plot(kind='bar', stacked=True, figsize=(7,7), title='Payment Method by Customer Type')
plt.ylabel("Count")
plt.show()
```

## Summary 

### Shape of the Data: The dataset contains 1,000 rows and 17 columns.
## Columns:
### Invoice ID: Unique identifier for each transaction.
### Branch: Branch code of the supermarket (e.g., A, B, or C).
### City: Location of the supermarket.
### Customer type: Whether the customer is a "Member" or "Normal."
### Gender: Gender of the customer.
### Product line: Category of the purchased product.
### Unit price: Price per unit of the product.
### Quantity: Number of units purchased.
### Tax 5%: 5% tax applied to the purchase.
### Total: Total amount including tax.
### Date: Date of purchase.
### Time: Time of purchase.
### Payment: Payment method (Cash, Credit card, Ewallet).
### COGS: Cost of goods sold.
### Gross margin percentage: Fixed at 4.76%.
### Gross income: Profit earned.
### Rating: Customer rating of the purchase (scale of 1-10).
        


## Dataset Overview:
### The dataset records transactional details from a supermarket across three branches in different cities (Yangon, Naypyitaw, Mandalay). It includes sales data like product line, unit price, quantity sold, and total sales amount, along with additional information such as customer type, gender, payment methods, gross income, and customer ratings.
